```python
import pandas as pd
import pandas.util.testing
pd.__version__
```

```python
#建立dict，匯入dataframe
dic = {
    "col 1": [1, 2, 3],
    "col 2": [10, 20, 30],
    "col 3": list('xyz'),
    "col 4": ['a', 'b', 'c'],
    "col 5": pd.Series(range(3))
}
df = pd.DataFrame(dic)
df
```

```python
#更變欄位名稱

#Way1
rename_col = {"col 1": "x", "col 2": "y"}
df.rename(rename_col, axis = 1)
#axis=1表columns，axis=0表row

#Way2
df.columns = ['xx','yy'] + list(df.columns[2:])
df
```

```python
#將已知分類型（categorical）欄位轉成（category）型態以節省記憶體。
#讀取特定欄位

dtypes = {"Embarked":"category",
         "Sex":"category"}
cols = ["PassengerId", "Name", "Sex", "Embarked"]

df = pd.read_csv('http://bit.ly/kaggletrain',
                 dtype = dtypes,
                 usecols = cols)
df.info(memory_usage="deep")
```

```python
#在有限記憶體讀取巨大csv檔案，可透過chunksize參數來限制一次讀取的資料筆數。
from IPython.display import display

reader = pd.read_csv('http://bit.ly/kaggletrain',
                    chunksize = 5,
                    usecols = cols)
for _, df_partial in zip(range(3), reader):
    display(df_partial)
```
